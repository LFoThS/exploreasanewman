# Define matrixs and make calculations
>>add and multiply included

```python
import numpy as np

arr_a = np.array([[12, 15, 16],
                  [1, 23, 25],
                  [14, 5, 6]])
arr_b = np.array([[56, 54, 85],
                  [45, 47, 69],
                  [41, 21, 32]])

print('sum of elements of a=', np.sum(arr_a))
print('sum of elements of b=', np.sum(arr_b))

print('a*b=\n', arr_a * arr_b)
print('np.dot(a, b)=\n', np.dot(arr_a, arr_b))

print('The transpose of a=\n', np.transpose(arr_a))
```
